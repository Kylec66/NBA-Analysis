```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
finals_rosters = "Resources/Finals_Teams_Payrolls.csv"

finals_payrolls_df = pd.read_csv(finals_rosters)

finals_payrolls_df
```

```python
# Total Payrolls of Both Teams
teams_df = finals_payrolls_df.groupby("Team")

teams_df

payrolls_df = teams_df["Salary"].sum()

payrolls_df

team_payrolls_df = pd.DataFrame({"Total Payroll": payrolls_df})

# team_payrolls_df["Total Payroll"] = team_payrolls_df["Total Payroll"].map("${:}".format)

team_payrolls_df
```

```python
# Average Age by Rosters

teams_df = finals_payrolls_df.groupby("Team")

teams_df

average_age_df = teams_df["Age"].mean()

average_age_df

age_comparison_df = pd.DataFrame({"Avg. Age": average_age_df})

# age_comparison_df["Avg. Age"] = age_comparison_df["Avg. Age"].map("{:.2f}".format)

age_comparison_df
```

```python
# Salary of Guards

guard_df = finals_payrolls_df[finals_payrolls_df['Pos'].str.startswith('G')]

guard_df

teams_df = guard_df.groupby("Team")

teams_df

guard_salary_df = teams_df["Salary"].sum()

guard_salary_df

guard_percent_df = guard_salary_df / payrolls_df * 100

guard_percent_df

team_guard_salaries_df =  pd.DataFrame({"Guards (G)": guard_salary_df,
                                       "% of Total Payroll (G)": guard_percent_df})

team_guard_salaries_df

#team_guard_salaries_df["Guards (G)"] = team_guard_salaries_df["Guards (G)"].map("${:}".format)
# team_guard_salaries_df["% of Total Payroll (G)"] = team_guard_salaries_df["% of Total Payroll (G)"].map("{:.2f}%".format)

team_guard_salaries_df
```

```python
# Average Age by Guard

guard_df = finals_payrolls_df[finals_payrolls_df['Pos'].str.startswith('G')]

guard_df

teams_df = guard_df.groupby("Team")

teams_df

guard_average_age_df = teams_df["Age"].mean()

guard_average_age_df

g_age_comparison_df = pd.DataFrame({"Avg. Age (G)": guard_average_age_df})

#g_age_comparison_df["Avg. Age (G)"] = g_age_comparison_df["Avg. Age (G)"].map("{:.2f}".format)

g_age_comparison_df
```

```python
# Salary of Forwards

forward_df = finals_payrolls_df[finals_payrolls_df['Pos'].str.startswith('F')]

forward_df

teams_df = forward_df.groupby("Team")

teams_df

forward_salary_df = teams_df["Salary"].sum()

forward_salary_df

forward_percent_df = forward_salary_df / payrolls_df * 100

forward_percent_df

team_forward_salaries_df =  pd.DataFrame({"Forwards (F)": forward_salary_df,
                                       "% of Total Payroll (F)": forward_percent_df})

team_forward_salaries_df

# team_forward_salaries_df["Forwards (F)"] = team_forward_salaries_df["Forwards (F)"].map("${:}".format)
# team_forward_salaries_df["% of Total Payroll (F)"] = team_forward_salaries_df["% of Total Payroll (F)"].map("{:.2f}%".format)

team_forward_salaries_df
```

```python
forward_df = finals_payrolls_df[finals_payrolls_df['Pos'].str.startswith('F')]

forward_df

teams_df = forward_df.groupby("Team")

teams_df

forward_average_age_df = teams_df["Age"].mean()

forward_average_age_df

f_age_comparison_df = pd.DataFrame({"Avg. Age (F)": forward_average_age_df})

#f_age_comparison_df["Avg. Age (F)"] = f_age_comparison_df["Avg. Age (F)"].map("{:.2f}".format)

f_age_comparison_df
```

```python
# Salary of Centers

center_df = finals_payrolls_df[finals_payrolls_df['Pos'].str.startswith('C')]

center_df

teams_df = center_df.groupby("Team")

teams_df

center_salary_df = teams_df["Salary"].sum()

center_salary_df

center_percent_df = center_salary_df / payrolls_df * 100

center_percent_df

team_center_salaries_df =  pd.DataFrame({"Centers (C)": center_salary_df,
                                       "% of Total Payroll (C)": center_percent_df})

team_center_salaries_df

# team_center_salaries_df["Centers (C)"] = team_center_salaries_df["Centers (C)"].map("${:}".format)
# team_center_salaries_df["% of Total Payroll (C)"] = team_center_salaries_df["% of Total Payroll (C)"].map("{:.2f}%".format)

team_center_salaries_df
```

```python
center_df = finals_payrolls_df[finals_payrolls_df['Pos'].str.startswith('C')]

center_df

teams_df = center_df.groupby("Team")

teams_df

center_average_age_df = teams_df["Age"].mean()

center_average_age_df

c_age_comparison_df = pd.DataFrame({"Avg. Age (C)": center_average_age_df})

#c_age_comparison_df["Avg. Age (C)"] = c_age_comparison_df["Avg. Age (C)"].map("{:.2f}".format)

c_age_comparison_df
```

```python
# Merge All DataFrames

teams_merge1_df = pd.merge(team_payrolls_df,age_comparison_df,on="Team")

teams_merge1_df
```

```python
teams_merge2_df = pd.merge(teams_merge1_df,team_guard_salaries_df,on="Team")

teams_merge2_df
```

```python
teams_merge3_df = pd.merge(teams_merge2_df,g_age_comparison_df,on="Team")

teams_merge3_df
```

```python
teams_merge4_df = pd.merge(teams_merge3_df,team_forward_salaries_df,on="Team")

teams_merge4_df
```

```python
teams_merge5_df = pd.merge(teams_merge4_df,f_age_comparison_df,on="Team")

teams_merge5_df
```

```python
teams_merge6_df = pd.merge(teams_merge5_df,team_center_salaries_df,on="Team")

teams_merge6_df
```

```python
team_full_merge_df = pd.merge(teams_merge6_df,c_age_comparison_df,on="Team")

team_full_merge_df
```

```python
# Create a group based on the values in the Position columns
position_group = team_full_merge_df.groupby('Team')

# Sum up the team's payrolls from the three positions
payroll_by_position = position_group['Guards (G)','Forwards (F)','Centers (C)'].sum()

payroll_by_position
```

```python
# Create a bar chart based off of the payroll by positon data
from matplotlib.colors import ListedColormap

colors = ListedColormap(['blue','red','purple'],name="from_list")

position_chart = payroll_by_position.plot(kind='bar',colormap = colors)

# Set the xlabel and ylabel using class methods
position_chart.set_xlabel("Team Positions")
position_chart.set_ylabel("Salary (x 10 million)")

# Create a Title and Set Y-Limit
plt.title("Total Payroll by Position")
plt.ylim(5000000, 85000000)

plt.savefig('positionpayroll.png')
plt.figure(figsize=(15,15))

plt.show()
```

```python
# Create a group based on the values in the Position columns
position_group = team_full_merge_df.groupby('Team')

# Sum up the team's payrolls from the three positions
position_percent = position_group['% of Total Payroll (G)','% of Total Payroll (F)','% of Total Payroll (C)'].sum()

position_percent
```

```python
# Make a variable called east_team and store a 'TOR' in it
east_team = "TOR"

# Collect the position percent of the 'east_team'
raptors_payroll_percent = position_percent.loc[east_team]

# Carry over Colormap
colors = ListedColormap(['blue','red','purple'],name="from_list")

# Create a pie chart based upon payroll position percent of 'east team'
raptors_payroll_percent = raptors_payroll_percent.plot(kind="pie",colormap = colors,y= ['% of Total Payroll (G)','% of Total Payroll (F)','% of Total Payroll (C)'],autopct='%.2f',title=("Payroll Position Percentage" ))

#raptors_payroll_percent.set_ylabel("Positon Percentage")

plt.savefig('raptors_payroll_percent_comp.png')
plt.figure(figsize=(15,15))

plt.show()
```
